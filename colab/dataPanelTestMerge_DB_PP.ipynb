Etude des genres

```python
genres_count = {}

# passage str en list
# ajout des genres en dico avec les count
def count_genres(genres):
  genres_list = genres.split(',') 
  for genre in genres_list:
    if genre in genres_count:
      genres_count[genre] +=1
    else :
      genres_count[genre] = 0
  return genres

#genres = 'Biography,Drama,Romance'
#count_genres(genres)
mergedFR['genres'].apply(count_genres)
```

```python
df_genres = pd.DataFrame.from_dict(
    genres_count,
    orient='index'
)
df_genres = df_genres.reset_index()
df_genres.columns=['Genres','Count']

df_genres = df_genres.sort_values(by='Count', ascending=False)
df_genres
```

```python
df_genres['Percentage'] = df_genres['Count']/565855*100
```

```python
df_genres
```

```python
GenreC = plt.gcf()
plt.figure(figsize=(8,6))
GenreC = sns.barplot(data=df_genres, x='Percentage',
                    y='Genres', orient="h")
plt.xlabel("Percentage of Films (%)")
plt.show()
```
